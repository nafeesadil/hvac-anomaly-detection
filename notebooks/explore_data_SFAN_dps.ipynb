```python
import pandas as pd
import matplotlib.pyplot as plt


# Load Excel file
file_path = '../data/hochschule_wismar_gebaeude_001_lueftung_hoersaele_750.xlsx'
df = pd.read_excel(file_path)

```

Below is the calculation of the percentage of the time the system was on per day. WEEKLY

```python
start_date = '2023-06-01'
end_date = '2023-06-07'
# Create fan status column
df['fan_on'] = df['SFAN_dps'] > 0

# Filter first month and create a safe copy
df = df[(df['datetime'] >= start_date) & (df['datetime'] < end_date)].copy()

# Now this won't raise a warning
df['date'] = df['datetime'].dt.date

# Compute daily ON ratio
daily_on_ratio = df.groupby('date')['fan_on'].mean() * 100


# Generate full date range index
all_dates = pd.date_range(start=start_date, end=end_date, freq='D').date

# Reindex to include all days, even missing ones (NaN)
daily_on_ratio = daily_on_ratio.reindex(all_dates)

# Plot
daily_on_ratio.plot(title='Fan ON Ratio per Day (%)', figsize=(12, 5), marker='o')
plt.ylabel("On Ratio (%)")
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
negative_values = df[df['SFAN_dps'] < 0]
print(negative_values[['datetime', 'SFAN_dps']])
```

```python
#HOURLY ON WEENEND (SATURDAY) THE PERCENTAGE OF TIME THE FAN WAS ON

import pandas as pd
import matplotlib.pyplot as plt

# Filter data for the specific date
day = '2023-06-2'
start = pd.to_datetime(day)
end = start + pd.Timedelta(days=1)

# Create a copy of the data for that day
day_df = df[(df['datetime'] >= start) & (df['datetime'] < end)].copy()

# Create fan_on column if not already present
day_df['fan_on'] = day_df['SFAN_dps'] > 0

# Extract the hour
day_df['hour'] = day_df['datetime'].dt.hour

# Group by hour and calculate how often the fan was ON
hourly_on_ratio = day_df.groupby('hour')['fan_on'].mean() * 100  # in %

# Plot
hourly_on_ratio.plot(kind='bar', title=f'Hourly Fan ON Ratio (%) on {day}', figsize=(10, 5))
plt.ylabel("Fan ON Ratio (%)")
plt.xlabel("Hour of Day")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
